### Type casting

```python
df['id'] = df['id'].astype(object)
df.dtypes
```

```python
df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

### What are the unique 'neighbourhood_group' values in New York?

```python
df['neighbourhood_group'].unique()
```

### How many listings are there in each 'neighbourhood_group'?

```python
df['neighbourhood_group'].value_counts()
```

### What is the average price of listings in each 'neighbourhood_group'?

```python
df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
```

### Find the top 5 most expensive listings

```python
df.nlargest(5, 'price')[['name', 'neighbourhood_group', 'price']]
```

### Filter listings that cost more than $500 and have over 50 reviews

```python
expensive_popular = df[(df['price'] > 500) & (df['number_of_reviews_ltm'] < 50)]
expensive_popular.head()
```

### Which 'room_type' is the most common?

```python
df['room_type'].value_counts()
```

### Find the listing with the maximum number of reviews

```python
max_rvw_idx = df['number_of_reviews'].idxmax()
max_rvw_idx
```

### Create a new column 'min_revenue' (price * minimum_nights)

```python
df['min_revenue'] = df['price'] * df['minimum_nights']
df[['name', 'price', 'minimum_nights', 'min_revenue']].head()
```

### Find the average price and minimum nights for each 'room_type'

```python
df.groupby('room_type').agg(avg_price = ('price', 'mean'), min_nights = ('minimum_nights' , 'min'))
```

### Top 10 hosts with the most listings

```python
df.groupby('host_name')['calculated_host_listings_count'].max().sort_values(ascending=False).head(10)
```

### Find all listings in a specific neighborhood, like 'Chelsea'

```python
chelsea_listings = df[df['neighbourhood'] == 'Chelsea']
print(f"Total listings in Chelsea: {len(chelsea_listings)}")
```

### Calculate the correlation between price, reviews, and availability

```python
df[['price', 'number_of_reviews', 'availability_365']].corr()
```

### Drop the 'license' column

```python
df.drop(columns=['license'], inplace = True)
```

### Find the cheapest listing in each 'neighbourhood_group'

```python
df.groupby('neighbourhood_group')['price'].min().sort_values(ascending = True)
```
